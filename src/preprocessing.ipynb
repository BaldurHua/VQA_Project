```python
import re

SENTENCE_SPLIT_REGEX = re.compile(r'(\W+)')

def tokenize(sentence):
    tokens = SENTENCE_SPLIT_REGEX.split(sentence.lower())
    tokens = [t.strip() for t in tokens if len(t.strip()) > 0]
    return tokens

def load_str_list(fname):
    with open(fname) as f:
        lines = f.readlines()
    lines = [l.strip() for l in lines]
    return lines


class VocabDict:

    def __init__(self, vocab_file):
        self.word_list = load_str_list(vocab_file)
        self.word2idx_dict = {w:n_w for n_w, w in enumerate(self.word_list)}
        self.vocab_size = len(self.word_list)
        self.unk2idx = self.word2idx_dict[''] if '' in self.word2idx_dict else None

    def idx2word(self, n_w):

        return self.word_list[n_w]

    def word2idx(self, w):
        if w in self.word2idx_dict:
            return self.word2idx_dict[w]
        elif self.unk2idx is not None:
            return self.unk2idx
        else:
            return self.word2idx_dict.get("<unk>", 0) 

    def tokenize_and_index(self, sentence):
        inds = [self.word2idx(w) for w in tokenize(sentence)]

        return inds
```

```python
vocab_questions_file = "C:/Users/Baldu/Desktop/Temp/VQA/data/vocab/vocab_questions.txt"
vocab_answers_file = "C:/Users/Baldu/Desktop/Temp/VQA/data/vocab/vocab_answers.txt"

# Create separate vocab dictionaries
vocab_questions = VocabDict(vocab_questions_file)
vocab_answers = VocabDict(vocab_answers_file)

print(f"Loaded {vocab_questions.vocab_size} words for questions")
print(f"Loaded {vocab_answers.vocab_size} words for answers")
```

```python
def is_multiple_choice(qa):
    raw_answers = [ans["answer"] for ans in qa["answers"]]
    return len(set(raw_answers)) > 1 
```

```python
def preprocess_for_pickle(data, vocab_q, vocab_a, img_dir, dataset_type, is_test=False):
    processed_data = []
    
    for qa in data:
        # print("Original question text:", qa["question"])
        
        tokenized_question = vocab_q.tokenize_and_index(qa["question"])
        # print("Tokenized question indices:", tokenized_question)  
        
        image_path = os.path.join(img_dir, f"COCO_{dataset_type}2014_{qa['image_id']:012d}.jpg")

        processed_entry = {
            "question_id": qa["question_id"],
            "image_id": qa["image_id"],
            "image_path": image_path,
            "question_tokens": tokenized_question,
            "is_multiple_choice": False if is_test else is_multiple_choice(qa),
        }

        if not is_test:
            raw_answers = [ans["answer"] for ans in qa["answers"]]
            tokenized_answers = [vocab_a.word2idx(ans) for ans in raw_answers if ans in vocab_a.word2idx_dict]

            if processed_entry["is_multiple_choice"]:
                processed_entry["valid_answers"] = tokenized_answers
            else:
                processed_entry["valid_answers"] = []

            if tokenized_answers:
                most_common_answer = Counter(tokenized_answers).most_common(1)[0][0]
                processed_entry["answer_label"] = most_common_answer
            else:
                processed_entry["answer_label"] = vocab_a.word2idx("<unk>")
        
        processed_data.append(processed_entry)

    return processed_data
```

```python
import pickle

train_img_dir = "C:/Users/Baldu/Desktop/Temp/VQA/data/images/train2014/"
val_img_dir = "C:/Users/Baldu/Desktop/Temp/VQA/data/images/val2014/"
test_img_dir = "C:/Users/Baldu/Desktop/Temp/VQA/data/images/test2015/"

train_processed = preprocess_for_pickle(merged_data,  vocab_questions, vocab_answers, train_img_dir, "train", is_test=False)
val_processed = preprocess_for_pickle(val_merged_data, vocab_questions, vocab_answers, val_img_dir, "val", is_test=False)
test_processed = preprocess_for_pickle(test_merged_data, vocab_questions, vocab_answers, test_img_dir, "test", is_test=True)


with open("C:/Users/Baldu/Desktop/Temp/VQA/data/preprocessed/preprocessed_train.pkl", "wb") as f:
    pickle.dump(train_processed, f)

with open("C:/Users/Baldu/Desktop/Temp/VQA/data/preprocessed/preprocessed_val.pkl", "wb") as f:
    pickle.dump(val_processed, f)

with open("C:/Users/Baldu/Desktop/Temp/VQA/data/preprocessed/preprocessed_test.pkl", "wb") as f:
    pickle.dump(test_processed, f)

print("Pickle files saved!")
```
